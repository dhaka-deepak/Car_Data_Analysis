```python
name = df["CarName"].head(20)
price = df["price"].head(20)

# Calculate the total price for each car name
total_price_by_name = price.groupby(name).sum()

# Create pie plot
fig, ax = plt.subplots()
ax.pie(total_price_by_name, labels=total_price_by_name.index, autopct='%1.1f%%',
        wedgeprops = {"edgecolor" : "black", #here we are providing borders around
                      'linewidth': 1,
                      'antialiased': True})

# Set aspect ratio to be equal, creating a circular pie plot
ax.set_aspect('equal')

# Set plot title
ax.set_title("Distribution of Car Prices by Name")

# Show the plot
plt.show()
```

```python
def twoSum(nums, target):
    a = []
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                a.append(i)
                a.append(j)
                break
            else:
                continue
    return a


# Driver code
nums = [2, 7, 11, 15]
target = 9

result = twoSum(nums, target)
print(result)
```
